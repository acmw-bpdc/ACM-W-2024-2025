<h3> Dimensionality Reduction </h3>
techniques to reduce the number of features (dimensions) in a dataset while retaining maximum information as possible. <br>
This is done for:
<li> reducing complexity
<li> improving model performance
<li> easier visualization
<br>
There are different types of methods for dimensionality reduction. <br>



| Type of Technique              | Dimensionality Reduction Techniques   |
|-------------------------------|---------------------------------------|
| **Linear Techniques**               | <li>Principal Component Analysis (PCA) <li>Linear Discriminant Analysis (LDA) <li>Independent Component Analysis (ICA) <li>Factor Analysis |
| **Non-Linear Techniques**           | <li>t-Distributed Stochastic Neighbor Embedding (t-SNE) <li>Uniform Manifold Approximation and Projection (UMAP) <li>Isomap <li>Locally Linear Embedding (LLE) <li>Multidimensional Scaling (MDS) <li>Spectral Embedding, <li>Kernel PCA |
| **Matrix Factorization Techniques** | <li>Non-Negative Matrix Factorization (NMF) <li>Singular Value Decomposition (SVD) |
| **Autoencoder-Based Techniques**    | <li>Autoencoders                                                   |

<h4> Clustering performance metrics </h4>
<ol>
<li> <b> Silhouette Score </b>
<ul>
<li> measure of how similar an object is to its own cluster
<li> range: -1 to 1
<li> higher values indicate better defined clusters
<li> negative values indicate data points are incorrectly clustered
</ul>
<li> <b> Davies-Bouldin Index </b>
<ul>
<li> measures clusters compactness and separation
<li> always greater than or equal to 0
<li> lower Davies-Bouldin scores indicate better defined clusters and higher values indicate overlapping
</ul>
<li> <b> Calinski-Harabasz Index </b>
<ul>
<li> evaluate quality of clusters by comparing the within-cluster variance and the between-cluster variance
<li> always greater than or equal to 0
<li> higher values indicate dense, well separated clusters while lower values suggest less clear clustering
</ul>
</ol>

<h4> K-means clustering </h4>
This algorithm divides the data into k (initially specified) clusters which can be represented by the mean (centroid) of the points in that cluster. <br>
It is a simple algorithm that works well for spherical, well separated clusters. However, it doesn't work very well with clusters of irregular shapes and requires k to be specified in advance.

```python
data = pd.read_csv('https://raw.githubusercontent.com/Shreyadoodipala/Unsupervised-ML-workshop/refs/heads/main/Customers.csv')
data.head()
```

```python
# Encode categorical data
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
data['Gender'] = le.fit_transform(data['Gender'])
data['Profession'] = le.fit_transform(data['Profession'])
```

```python
# Normalize numeric data
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data[['Age', 'Annual Income ($)', 'Spending Score (1-100)', 'Work Experience', 'Family Size']])
df_scaled = pd.DataFrame(scaled_data, columns=['Age', 'Annual Income ($)', 'Spending Score (1-100)', 'Work Experience', 'Family Size'])
```

```python
# Append the columns
final_data = pd.concat([data[['Gender', 'Profession']].reset_index(drop=True), df_scaled.reset_index(drop=True)], axis=1)
```

inertia or sum of squared errors (SSE) - measure of the compactness of the clusters <br>
**Elbow method**

```python
sse = []
k_values = range(2, 16)

for k in k_values:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(final_data)
    sse.append(kmeans.inertia_)
```

```python
# Plotting the results
plt.plot(k_values, sse, 'bo-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
kmeans_final = KMeans(n_clusters=9)
kmeans_final.fit(final_data)
```

```python
# Visualize the clusters

# Initialize PCA with the desired number of components
pca = PCA(n_components=2)  # Reduce to 2 dimensions

# Fit and transform the data
data_pca = pca.fit_transform(final_data)

df_pca = pd.DataFrame({
    'PCA1': data_pca[:, 0],
    'PCA2': data_pca[:, 1],
    'Cluster': kmeans_final.labels_
})

# 2D scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', palette='viridis', data=df_pca, s=100, edgecolor='k')

# Setting labels
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.title('2D PCA Visualization')

plt.grid(True)
plt.show()
```

```python
# Fit and transform the data using PCA to 3 dimensions
pca = PCA(n_components=3)
data_pca = pca.fit_transform(final_data)

# Create a DataFrame with the PCA components and cluster labels
df_pca = pd.DataFrame({
    'PCA1': data_pca[:, 0],
    'PCA2': data_pca[:, 1],
    'PCA3': data_pca[:, 2],
    'Cluster': kmeans_final.labels_  # Add the cluster labels
})

# 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Create the 3D scatter plot
scatter = ax.scatter(df_pca['PCA1'], df_pca['PCA2'], df_pca['PCA3'], c=df_pca['Cluster'], cmap='viridis',
                     s=50, edgecolor='k')

# Add axis labels
ax.set_xlabel('Component 1')
ax.set_ylabel('Component 2')
ax.set_zlabel('Component 3')
plt.title('3D PCA Visualization of K-Means Clusters')

# Add a legend or color bar for cluster labeling
plt.colorbar(scatter)

plt.show()
```

```python
silhouette_kmeans = silhouette_score(final_data, kmeans_final.labels_)
print("Silhouette Score: ", silhouette_kmeans)

davies_bouldin_kmeans = davies_bouldin_score(final_data, kmeans_final.labels_)
print("Davies-Bouldin Score: ", davies_bouldin_kmeans)

calinski_harabasz_kmeans = calinski_harabasz_score(final_data, kmeans_final.labels_)
print("Calinski-Harabasz Score: ", calinski_harabasz_kmeans)
```

<h4> DBSCAN </h4>
Density Based Spatial Clustering of Applications with Noise is very good for clusters with arbitrary shape and detecting noise and outliers. <br>
However, there are 2 parameters 'eps' and 'min_samples' that need to be set for this method. The result of DBSCAN is highly sensitive to these parameters.
<br><br>
<b>eps: </b>Epsilon specifies the maximum distance between two samples for one to be considered as in the neighborhood of the other. <br>
<b>min_samples: </b> This specifies the minimum number of samples (including the point itself) that must be within eps distance to form a dense region.

```python
from sklearn.neighbors import NearestNeighbors
```

```python
# Use k-nearest neighbors to find the distances
neighbors = NearestNeighbors(n_neighbors=5)  # k = 5
neighbors_fit = neighbors.fit(final_data)
distances, indices = neighbors_fit.kneighbors(final_data)
```

```python
# Sort the distances and plot them
distances = np.sort(distances[:, 4])  # We take the 4th neighbor distance (index 4)
plt.plot(distances)
plt.ylabel("4th Nearest Neighbor Distance")
plt.xlabel("Data Points sorted by distance")
plt.title("k-NN Distance Plot for Selecting eps")
plt.grid(True)
plt.show()
```

```python
# Apply DBSCAN
# Set eps (distance threshold) and min_samples (minimum points to form a dense region)
dbscan = DBSCAN(eps=1.5, min_samples=5)
dbscan.fit(final_data)
dbscan_labels = dbscan.labels_ # get cluster variables
```

```python
# Perform PCA to reduce to 2 dimensions for visualization
pca = PCA(n_components=2)
data_pca = pca.fit_transform(final_data)

# Create a DataFrame with PCA components and cluster labels
df_pca = pd.DataFrame({
    'PCA1': data_pca[:, 0],
    'PCA2': data_pca[:, 1],
    'Cluster': dbscan_labels
})
```

```python
# Plot the clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', palette='viridis', data=df_pca, s=100, edgecolor='k')

plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.title('2D PCA Visualization of DBSCAN Clustering')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
# Fit and transform the data using PCA to 3 dimensions
pca = PCA(n_components=3)
data_pca = pca.fit_transform(final_data)

# Create a DataFrame with the PCA components and cluster labels
df_pca = pd.DataFrame({
    'PCA1': data_pca[:, 0],
    'PCA2': data_pca[:, 1],
    'PCA3': data_pca[:, 2],
    'Cluster': dbscan_labels  # Add the cluster labels
})

# 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Create the 3D scatter plot
scatter = ax.scatter(df_pca['PCA1'], df_pca['PCA2'], df_pca['PCA3'], c=df_pca['Cluster'], cmap='viridis',
                     s=50, edgecolor='k')

# Add axis labels
ax.set_xlabel('Component 1')
ax.set_ylabel('Component 2')
ax.set_zlabel('Component 3')
plt.title('3D PCA Visualization of DBSCAN Clusters')

# Add a legend or color bar for cluster labeling
plt.colorbar(scatter)

plt.show()
```

```python
silhouette_dbscan = silhouette_score(final_data, dbscan_labels)
print("Silhouette Score: ", silhouette_dbscan)

davies_bouldin_dbscan = davies_bouldin_score(final_data, dbscan_labels)
print("Davies-Bouldin Score: ", davies_bouldin_dbscan)

calinski_harabasz_dbscan = calinski_harabasz_score(final_data, dbscan_labels)
print("Calinski-Harabasz Score: ", calinski_harabasz_dbscan)
```

<h3> Anomaly Detection </h3>
identifying data or events that are outside the normal range

```python
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']

df_diabetes = pd.read_csv(url, names=columns)
```

```python
df_diabetes.head()
```

```python
# Separate features and target
X = df_diabetes.drop(columns='Outcome')
y = df_diabetes['Outcome']
```

```python
# Standardize the features
scaler_anomaly = StandardScaler()
X_scaled = scaler_anomaly.fit_transform(X)
```

```python
from sklearn.ensemble import IsolationForest
```

```python
# Create and fit the model
isof = IsolationForest(contamination=0.34)
isof.fit(X_scaled)

# Predict: 1 for normal, -1 for outlier
y_pred_isof = isof.predict(X_scaled)
# Map the predictions to original labels (-1 for anomaly, 1 for normal)
y_pred_isof = [1 if x == -1 else 0 for x in y_pred_isof]
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Evaluate the results
print(confusion_matrix(y, y_pred_isof))
print(classification_report(y, y_pred_isof))
```

```python
print(y_pred_isof[:10])
```

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
lof = LocalOutlierFactor(contamination=0.35)
lof.fit(X_scaled)
y_pred_lof = lof.fit_predict(X_scaled)
y_pred_lof = [1 if x == -1 else 0 for x in y_pred_lof]
```

```python
# Evaluate the results
print(confusion_matrix(y, y_pred_lof))
print(classification_report(y, y_pred_lof))
```

<h3> Association Rule Mining </h3>
It's used to find relationships (associations) or patterns among sets of items in a large dataset. <br>
The most common application is market basket analysis i.e. understand what products are often bought together. Other applications include suggesting users additional products to buy, associations between symptoms and diseases, and between different drugs and side effects. <br>


<h6> Itemset </h6>
A list of one or more items.

<h6> Rule </h6>
{milk}  {bread} <br>
This means that whenever milk is purchased, bread is also likely to be purchased in the same transaction.

<h6> Antecedents and Consequents </h6>
Antecedents are item(s) on the left-hand side of the rule, while consequents are item(s) on the right-hand side of the rule.

<h6> Support </h6>
The proportion of transactions containing an itemset X.
$$\text{Support(X)} = \frac{\text{Number of transactions containing X}}{\text{Total number of transactions}}$$

<h6> Confidence </h6>
It indicates how good a rule is. It measures the likelihood that a transaction containing an antecedent also contains the consequent.
$$\text{Confidence(Antecedent} \rightarrow \text{Consequent)} = \frac{\text{Support(Antecedent} \cap \text{Consequent)}}{\text{Support(Antecedent)}}$$

<h6> Lift </h6>
It indicates whether the antecedent and consequent are independent of each other or not.
$$\text{Lift(Antecedent} \rightarrow \text{Consequent)} = \frac{\text{Confidence(Antecedent} \rightarrow \text{Consequent)}}{\text{Support(Consequent)}}$$
<ul>
<li> Lift = 1: Antecedent and consequent are independent (no association).
<li> Lift > 1: Positive association (the presence of the antecedent increases the likelihood of the consequent).
<li> Lift < 1: Negative association (the presence of the antecedent decreases the likelihood of the consequent).
</ul>


```python
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
# Load the dataset
url = "https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/groceries.csv"
```

```python
# Read the file line by line and split items
transactions = []

# Using the requests library to read the CSV content
import requests
response = requests.get(url)
lines = response.text.splitlines()

for line in lines:
    items = line.split(',')
    transactions.append(items)
```

```python
# Determine the maximum number of items in any transaction
max_items = max(len(transaction) for transaction in transactions)

# Create a DataFrame, filling shorter rows with NaN
data = []

for transaction in transactions:
    # Pad with NaN for shorter transactions
    padded_transaction = transaction + [np.nan] * (max_items - len(transaction))
    data.append(padded_transaction)

# Create the DataFrame with NaN for missing values
df = pd.DataFrame(data)

# Display the first few rows
df.head()
```

```python
# Convert DataFrame to a list of lists, dropping NaN values
transactions_list = df.apply(lambda x: x.dropna().tolist(), axis=1).tolist()
```

```python
# One-hot encode the transactions
te = TransactionEncoder()
te_array = te.fit(transactions_list).transform(transactions_list)
df_onehot = pd.DataFrame(te_array, columns=te.columns_)
```

```python
# Apply the Apriori algorithm with a minimum support threshold of 0.01
frequent_itemsets = apriori(df_onehot, min_support=0.01, use_colnames=True)

# Generate association rules with a minimum confidence threshold of 0.5
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)

# Display the rules
rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
```
